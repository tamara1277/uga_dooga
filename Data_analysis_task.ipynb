# Required imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import json
from ipywidgets import interact
import plotly.graph_objects as go
```

# Data preparation

```python
data = pd.read_csv('us events ad_revenue filtered 03.02-07.02.csv')
data.head()
```

```python
data_json = pd.DataFrame(data['event_json'].apply((json.loads)).values.tolist())
data_json.head()
```

```python
data = pd.concat([data.drop('event_json', axis=1), data_json], axis=1)
data.head()
```

```python
data.info()
```

```python
data['city'] = data['city'].replace(np.nan, 'missing')
```

# Functions

```python
def data_grouping(data, group):
    df = data.groupby(["ad_type",group])['revenue'].apply(np.mean)
    df = df.to_frame().reset_index().sort_values(by="revenue", ascending=False)
    
    df["revenue"] = (df["revenue"]*1000)
    df = df.rename(columns={"revenue": "eCPM"})
    
    return df
```

```python
def interactive_visualizations(data, col_name, graph_type, y_value):
    if graph_type == 'hist':
        def hist(chooser):
            for i in col_name:
                if chooser == i:
                    fig = px.histogram(data[data['ad_type'] == i],
                                       x=y_value, nbins=100)
                    fig.show()
            return
        print(interact(hist, chooser = col_name))

    elif graph_type == 'table':
        def table(chooser):
            for i in col_name:
                if chooser == i:
                    print(data[i].value_counts())
            return
        print(interact(table, chooser = col_name))
    
    elif graph_type == 'bar':
        def series(chooser):
            for i in col_name[:-1]:
                if chooser == i:
                    fig = px.bar(data[int(i.split()[1]) - 1 :int(i.split()[-1])],
                                 x='city', y=y_value)
                    fig.show()
                    
            return
        print(interact(series, chooser = col_name[:-1]))
    
    elif graph_type == 'os_bar':
        def mean_os(chooser):
            for i in col_name:
                if chooser == i:
                    fig = px.bar(data[data['ad_type'] == i],
                                 x='os_version',
                                 y=y_value)
                    fig.show()
            return
        print(interact(mean_os, chooser = col_name))
    
        
    
```

```python
def conditions(event_datetime):
    event_datetime = int(event_datetime.split()[1].split(':')[0])
    
    if event_datetime >= 5 and event_datetime < 12:
        return "morning"
    elif event_datetime >= 12 and event_datetime < 17:
        return "afternoon"
    elif event_datetime >= 17 and event_datetime < 21:
        return "evening"
    else:
        return "night"
```

## Task 2 

With 95% confidence level and 5% error confidence interval, for given population
the minimum required sample size equals to 363. Based on the latter, we may remove
cities which have less than 363 rows of data.

Afterwards, we will check again the average eCPM metric for each ad type per countries.

```python
filtered_data = data.groupby(['ad_type', 'city']).agg({'profile_id': ['count'], 'revenue':['mean']})
filtered_data[('revenue', 'mean')] *= 1000

filtered_data = filtered_data[filtered_data[('profile_id', 'count')] >= 363
                             ].reset_index().sort_values(
                                       by=('revenue', 'mean'), ascending=False)

filtered_data.columns = filtered_data.columns.droplevel(1)
filtered_data = filtered_data.rename(columns={'profile_id': "count",
                                              'revenue': "eCPM"})
```

```python
col_name_city = []
ad_type = ["banner","int","rv"]
for j in ad_type:
    l = []
    for i in range(0, len(filtered_data[filtered_data['ad_type'] == j]) , 20):
        l.append(f'Top {i+1} - {i+20}')
    col_name_city.append(l)
```

### Banner 

```python
interactive_visualizations(filtered_data[filtered_data['ad_type'] == 'banner'], 
                           col_name_city[0], 'bar', 'eCPM')
```

### INT 

```python
fig = px.bar(filtered_data[filtered_data['ad_type'] == 'int'],
                                 x='city', y='eCPM')
fig.show()
```

### RV 

```python
fig = px.bar(filtered_data[filtered_data['ad_type'] == 'rv'],
                                 x='city', y='eCPM')
fig.show()
```

# Task 4.2

Does the eCPM metric depend on the time of the day?

```python
time_data = data[['event_datetime', 'revenue']]
time_data.head()
```

```python
time_list = []
for i in time_data['event_datetime']:
    time_list.append(conditions(i))
    
time_data.insert(loc=2, column='time_category', value=time_list)

time_data.head()
```

```python
mean_time_category = time_data.groupby(["time_category"])['revenue'].mean()
mean_time_category = pd.DataFrame(mean_time_category * 1000).sort_values(by="revenue",
                                                              ascending=False)
mean_time_category
```

```python
labels = mean_time_category.index
values = mean_time_category['revenue']

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

                               Data insights
                               
Overall, revenue does not have strong dependency from the time of day, though the highest revenue is generated in the evening & morning.
